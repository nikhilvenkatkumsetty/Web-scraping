```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import urllib.parse
from urllib.parse import urljoin
import plotly.graph_objects as go
import plotly.express as px
import re
```

```python
urlmain = "https://www.tripadvisor.com/Hotels-g304551-New_Delhi_National_Capital_Territory_of_Delhi-Hotels.html"
base_url = "https://www.tripadvisor.com"
```

```python
list_hotel = []
full_lst = []
local_lst = []
```

```python
html = requests.get(urlmain)
soup = BeautifulSoup(html.content, 'lxml')
```

```python
for item in soup.find_all("a", attrs={"class":"respListingPhoto"}):
    full_loc_url = []
    loc_url = item.get("href")
    full_loc_url = urljoin(base_url,loc_url)
    local_lst.append(full_loc_url)
```

```python
urlhead = "https://www.tripadvisor.com/Hotels-g304551"
urltail = "-New_Delhi_National_Capital_Territory_of_Delhi-Hotels.html"
```

```python
Hotel_Name = []
Price = []
Amenities = []
Reviews = []
Photos = []
```

```python
hotel_html = pd.Series(local_lst).drop_duplicates().tolist()
```

```python
len(hotel_html)
```

```python
for url in hotel_html:
    link = requests.get(url)
    soup = BeautifulSoup(link.content, 'lxml')
    try:
        nm = soup.find("h1", attrs={"id":"HEADING"})
        Hotel_Name.append(nm.text)
    except:
        Hotel_Name.append("NA")
    try:
        pc = soup.find("div", attrs={"data-sizegroup":"hr_chevron_prices"})
        Price.append(pc.text)
    except:
        try:
            pc = soup.find("div", attrs={"class":"hotels-hotel-offers-DominantOffer__price--D-ycN"})
            Price.append(pc.text)
        except:
            Price.append("NA")
    try:
        s1=""
        for am in soup.find_all("div", class_="_2rdvbNSg"):
            if am.get_text():
                s1+=+am.get_text()+", "        
        Amenities.append(s1)
    except:
        Amenities.append("NA")
    try:
        r = soup.find("span", attrs={"class":"_3cjYfwwQ"})
        Reviews.append(r.text)
    except:
        Reviews.append("NA")
    try:
        l2=[]
        for p in soup.find_all("img", class_="_1a4WY7aS"):
            l2.append(p.get('src'))        
        Photos.append(l2)
    except:
        Photos.append("NA")
```

```python
print(Hotel_Name)
print(Price)
print(Amenities)
print(Reviews)
print(Photos)
```

```python
print(len(Hotel_Name))
print(len(Price))
print(len(Amenities))
print(len(Reviews))
print(len(Photos))
```

```python
df1 = pd.DataFrame(Hotel_Name)
df2 = pd.DataFrame(Price)
df3 = pd.DataFrame(Amenities)
df4 = pd.DataFrame(Reviews)
df5 = pd.DataFrame(Photos)
```
